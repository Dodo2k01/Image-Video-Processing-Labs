#### Step 2: Brightness Adjustment

Use `cv2.convertScaleAbs()` to perform brightness adjustment. First, increase the brightness of the original image by `70` to obtain a `brightimage`. Then decrease the brightness of the original image by `70` to obtain a `darkimage`. Display the result using matplotlib in a 1x3 grid with original, bright and dark images.

```python
# Function to adjust brightness
def adjust_brightness(image, value=30):
    adjusted = cv2.convertScaleAbs(image, alpha=1, beta = value)
    return adjusted

# Generate a bright_image with 70 more brightness
bright_image = adjust_brightness(image, 70)

#Generate a dark_image with 70 less brightness
dark_image = adjust_brightness(image, -70)

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display bright image
plt.subplot(1, 3, 2)
plt.imshow(bright_image, cmap='gray')
plt.title("Brighter")
plt.axis("off")

# Display dark  image
plt.subplot(1, 3, 3)
plt.imshow(dark_image, cmap='gray')
plt.title("Darker")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#### Step 3: Contrast Adjustment

Use `cv2.convertScaleAbs()` to perform contrast adjustment. First increase the contrast of the original image by `1.8` to obtain a `contrastimage`. Then decrease the contrast of the original image by `0.1` to obtain a `flatimage`. Display the result using matplotlib in a 1x3 grid with original, contrast and flat images.

```python
# Function to adjust brightness
def adjust_contrast(image, factor=1.5):
    adjusted = cv2.convertScaleAbs(image, alpha=factor, beta=0)
    return adjusted

# Generate a contrast_image with 1.8 more contrast
contrast_image = adjust_contrast(image, 1.8)

#Generate a flat_image with 0.1 less contrast
flat_image = adjust_contrast(image, 0.1)

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display contrast image
plt.subplot(1, 3, 2)
plt.imshow(contrast_image, cmap='gray')
plt.title("Contrast")
plt.axis("off")

# Display flat image
plt.subplot(1, 3, 3)
plt.imshow(flat_image, cmap='gray')
plt.title("Flat")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#### Step 4: Thresholding

Use `cv2.threshold()` to perform thresholding. First turn all the values below `128` to `black` but keep the rest as it is to obtain the image `thresholded`. Then turn all values below `128` to `black` and all values above `128` to `white` to obtain the image `binary`. Both of these operations needs to be executed by the same function. Display the result using matplotlib in a 1x3 grid with original, thresholded and binary images.

```python
# Function to apply thresholding
def apply_threshold(image, type = "zero", threshold_value=128):
    if type== "zero":
        _, thresh_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_TOZERO)
    elif type == "binary":
        _, thresh_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
    else: #no change
        thresh_image = image
    return thresh_image

thresholded = apply_threshold(image)
binary = apply_threshold(image, "binary")

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display thresholded image
plt.subplot(1, 3, 2)
plt.imshow(thresholded, cmap='gray')
plt.title("thresholded")
plt.axis("off")

# Display binary image
plt.subplot(1, 3, 3)
plt.imshow(binary, cmap='gray')
plt.title("Binary")
plt.axis("off")

plt.tight_layout()
plt.show()
```

### Part 2 Neighborhood (Local) Operations

#### Step 1: Smoothing
Write a function that performs smoothing based on the `type` specified. For mean smoothing use `cv2.filter2D()` and for Gaussian smoothing use `cv2.GaussianBlur()`. Display the result using matplotlib in a 1x3 grid with original, mean blurred and Gaussian blurred images.

```python
def blur_image(image, type="mean",size=3):
    if type=="mean":
        kernel = np.ones((size,size), np.float32)/(size**size)
        smoothed_image = cv2.filter2D(image, ddepth=-1, kernel=kernel)
        return smoothed_image
    elif type == "gaussian":
        smoothed_image = cv2.GaussianBlur(image, ksize=(size,size), sigmaX=1)
        return smoothed_image
    return image

mean_blur     = blur_image(image,type="mean",size=5)
gaussian_blur = blur_image(image,type="gaussian",size=7)

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display mean image
plt.subplot(1, 3, 2)
plt.imshow(mean_blur, cmap='gray')
plt.title("Mean")
plt.axis("off")

# Display gaussian image
plt.subplot(1, 3, 3)
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#### Step 2: Edge Detection
Write a function that performs edge detection based on the `type` specified. Use `cv2.filter2D()` to apply the following kernels: `Prewitt-X`, `Sobel-Y` and `Laplacian-8`. Display the result using matplotlib in a 1x4 grid with original, prewitt-x, sobel-y, and laplacian-8 images.

```python
def edge_detection(image, type):
    if type=="prewitt_x":
        kernel = np.array([[-1, 0, 1],
                          [-1, 0, 1],
                          [-1, 0, 1]], dtype=np.float32)

    elif type=="sobel_y":
        kernel = np.array([[-1, -2, -1],
                           [0, 0, 0],
                           [1, 2, 1]], dtype=np.float32)
    else:
        #laplacian8
        kernel = np.array([[-1, -1, -1],
                           [-1, 8, -1],
                           [-1, -1, -1]], dtype=np.float32)

    edges = cv2.filter2D(image, -1, kernel)
    return edges

prewitt_x   = edge_detection(image, 'prewitt_x')
sobel_y     = edge_detection(image, 'sobel_y')
laplacian_8 = edge_detection(image, 'laplacian_8')

Lab = True
# Display the results
plt.figure(figsize=(8, 6))

if Lab:
    # Original Image
    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")
    
    # Prewitt-X
    plt.subplot(1, 4, 2)
    plt.imshow(prewitt_x, cmap='gray')
    plt.title("Prewitt-X")
    plt.axis("off")

    # Sobel-Y
    plt.subplot(1, 4, 3)
    plt.imshow(sobel_y, cmap='gray')
    plt.title("Sobel-Y")
    plt.axis("off")

    # Laplacian-8
    plt.subplot(1, 4, 4)
    plt.imshow(laplacian_8, cmap='gray')
    plt.title("Laplacian-8")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

#### Step 3: Sharpening
Write a function that performs sharpening based on the `type` specified. Use `cv2.filter2D()` to apply the following kernels: `Laplacian-8` and a basic sharpening kernel from the lecture notes. Display the result using matplotlib in a 1x3 grid with original, laplacian and kernel sharpened images.

```python
def sharpen(image, type):
    if type=='laplacian':
        kernel_sharpen = np.array([[-1,-1,-1],
                                   [-1, 8, -1],
                                   [-1, -1, -1]], dtype=np.float32)
        sharpened = cv2.filter2D(image, ddepth=-1, kernel=kernel_sharpen)
        sharpened = cv2.add(image, sharpened)
    else:
        kernel_sharpen = np.array([[0 ,-1, 0],
                                   [-1, 5, -1],
                                   [0, -1, 0]], dtype=np.float32)
        sharpened = cv2.filter2D(image, ddepth=-1, kernel=kernel_sharpen)
    return sharpened


laplacian_sharpen = sharpen(image,type="laplacian")
kernel_sharpen    = sharpen(image,type="kernel")

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display kernel image
plt.subplot(1, 3, 2)
plt.imshow(kernel_sharpen, cmap='gray')
plt.title("Kernel")
plt.axis("off")

# Display laplacian sharpening image
plt.subplot(1, 3, 3)
plt.imshow(laplacian_sharpen, cmap='gray')
plt.title("Laplacian")
plt.axis("off")

plt.tight_layout()
plt.show()
```

### Part 3 Morphological Operations
#### Step 1: Structuring Elements
Use `cv2.getStructuringElement` to create square, circular and cross-shaped Structuring Elements or define manually as a 5x5 neighbourhood.

```python
def get_structuring_element(type):
    if type=="square":
        # Square structuring element

        se = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
    elif type=="circle":
        # Circular structuring element
        se = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    else:
        # Cross structuring element
        se = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
    return se

square = get_structuring_element("square")
circle = get_structuring_element("circle")
cross  = get_structuring_element("cross")
```

#### Step 2: Erosion
Use `cv2.erode` to perform erosion with all three previously defined Structuring Elements

#### Step 2: Translation
Use `cv2.warpAffine` to move the image on the x and y axes. Display the result using matplotlib in a 1x3 gridimages.

```python
def translate_image(image, t_x, t_y):

    #Define the translation matrix
    translation_matrix = np.array([[1, 0, t_x], [0, 1, t_y]], dtype=np.float32)

    # Perform the translation
    rows, cols = image.shape
    translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))
    
    return translated_image

translate_1 = translate_image(image,50,30)
translate_2 = translate_image(image,-50,-30)

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display translate-1 image
plt.subplot(1, 3, 2)
plt.imshow(translate_1, cmap='gray')
plt.title("Translate-1")
plt.axis("off")

# Display translate-2 image
plt.subplot(1, 3, 3)
plt.imshow(translate_2, cmap='gray')
plt.title("Translate-2")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#### Step 3: Rotation
Use `cv2.warpAffine` to rotate the image with the given angle on its center. Display the result using matplotlib in a 1x3 gridimages.

```python
def rotate_image(image, angle):
    # Get the dimensions of the image
    h, w = image.shape
    # Define the rotation center as the center of the image and rotation angle
    center = (h//2, w//2)
    # Create the rotation matrix and apply it
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    # np.array([[np.cos(np.radians(angle)), -np.sin(np.radians(angle))], [np.sin(np.radians(angle)), np.cos(np.radians(angle))]], dtype=np.float32)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (w,h))

    return rotated_image

rotate_1 = rotate_image(image,45)
rotate_2 = rotate_image(image,-30)

# Display the results in a 1x3 grid
plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Display rotate-1 image
plt.subplot(1, 3, 2)
plt.imshow(rotate_1, cmap='gray')
plt.title("Rotate-1")
plt.axis("off")

# Display rotate-2 image
plt.subplot(1, 3, 3)
plt.imshow(rotate_2, cmap='gray')
plt.title("Rotate-2")
plt.axis("off")

plt.tight_layout()
plt.show()
```
